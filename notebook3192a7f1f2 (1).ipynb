```python
Y=train_data['label']
X=train_data.drop('label',axis=1)
```

```python
print(X.shape,Y.shape)
```

```python
import nltk
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
stopwords=stopwords.words('english')
stopwords
```

```python
text=X.copy()
print(text['title'][1])
text.reset_index(inplace=True)
print(text['title'][1])
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
vocab_size=5000
maxlen=100
def data_preprocess(texts):
    result=[]
    for i in range(0,len(texts)):
        pre_text = re.sub('[^a-zA-Z]', ' ', str(texts['title'][i]))
        pre_text = pre_text.lower()
        pre_text = pre_text.split()
    
        pre_text = [PorterStemmer().stem(word) for word in pre_text if not word in stopwords]
        pre_text = ' '.join(pre_text)        
        result.append(pre_text)

    one_hot_rep=[one_hot(words,vocab_size) for words in result]
    embedded_text=pad_sequences(one_hot_rep,padding='pre',maxlen=maxlen)
    return embedded_text
```

```python
embedded_text=data_preprocess(text)
X=np.array(embedded_text)
Y=np.array(Y)    
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout, Bidirectional ,LayerNormalization
```

```python
model=Sequential()
model.add(Embedding(input_dim=vocab_size,output_dim=254,input_length=maxlen))
model.add(Bidirectional(LSTM(254)))
model.add(LayerNormalization())
model.add(Dropout(0.2))
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(64,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2)
```

```python
model.fit(X_train,Y_train,validation_data=(X_val,Y_val),epochs=16,batch_size=128)
```

```python
from sklearn.metrics import accuracy_score
Y_pred=model.predict_classes(X_val)
accuracy_score(Y_val,Y_pred)
```

```python
test_text=test_data.copy()
test_text = test_text.fillna(' ')
test_text.reset_index(inplace=True)
test_embedd=data_preprocess(test_text)
test_array=np.array(test_embedd)
test_pred=model.predict_classes(test_array)
test_pred=test_pred.ravel()
```

```python
submission = pd.DataFrame({'id':test_text['id'], 'label':test_pred})
print(submission.shape)
submission.head(5)
```
